# Table 3 — Adjuster Profile and Capability
## Purpose
Reference table for all 15 adjusters in the system. Contains role, licensing,
territory, authority limits, specialisations, and current active claim count.

## Key design decisions
- **Authority Limits** — set by role seniority. Adjuster I < II < Specialist < Supervisor.
- **LOB Eligibility** — each adjuster is restricted to the lines of business their
  role and licensing covers. Multi-line eligibility is reserved for senior roles.
- **Regions/Jurisdiction Authorization** — reflects realistic state licensing clusters
  (e.g. WV/VA/MD together, NY/NJ/CT/PA together).
- **Active Claim Count** — hand-set to reflect realistic load differences across roles.
  CAT specialists carry the highest counts (82) during active deployments;
  SIU leads carry the lowest (15) due to case complexity.
- **Attributes** — industry certifications: Xactimate (property estimating),
  HAAG (hail/wind), ASE/Audatex (auto damage), CWCP (workers comp), CFI (fraud),
  CPCU/SCLA (senior designations).
